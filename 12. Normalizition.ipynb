```python
from sklearn import preprocessing
import pandas as pd
```

```python
PATIENT_FILE_PATH = 'Downloads/liver_patient_data.csv'
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
liver_patients_df = pd.read_csv(PATIENT_FILE_PATH)
```

```python
liver_patients_df.describe()
```

```python
df = liver_patients_df[['Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase']]
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
nomalized_data = scaler.fit_transform(df)
```

```python
nomalized_df = pd.DataFrame(nomalized_data, columns = ['Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase'])
```
